## Complexidade Exponencial

Exemplo: Fibonacci

Cada execução gera mais n execuções necessárias

Exemplo:
https://www.geeksforgeeks.org/python-program-to-print-the-fibonacci-sequence/?ref=header_search

Provocação: uma implementação ótima de fibonacci tem Complexidade Linear, como?

```python
def fib(n):
  if (n <= 1):
    return n
  else
    return fib(n-1) + fib(n+1)
```
